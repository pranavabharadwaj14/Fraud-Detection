```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Load the data
data = pd.read_excel('book1.xlsx')  # Replace with the actual path to your dataset

data.head()
```

```python
data=data.drop(columns=['nameOrig'])
data=data.drop(columns=['newbalanceDest'])
data=data.drop(columns=['oldbalanceDest'])
data=data.drop(columns=['step'])
data=data.drop(columns=['nameDest'])
```

```python
data.head()
```

```python
data=data.drop(columns=['isFlaggedFraud'])
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Define columns to be label encoded
cols_to_label_encode = ['type']

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Iterate over each column and apply label encoding
for col in cols_to_label_encode:
    data[col] = label_encoder.fit_transform(data[col])
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(columns=['isFraud'])  # Features
y = data['isFraud']  # Target variable

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Print shapes to verify
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

# Fit on training data and transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
```

```python
logreg_model = LogisticRegression()
```

```python
logreg_model.fit(X_train_scaled, y_train)
```

```python
y_pred = logreg_model.predict(X_test_scaled)
y_pred_proba = logreg_model.predict_proba(X_test_scaled)[:, 1]
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import joblib
joblib.dump(logreg_model, 'logreg_model.pkl')
print("Model saved as logreg_model.pkl")
```
